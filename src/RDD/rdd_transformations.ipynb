```python
from pyspark import SparkContext, SparkConf
```

```python
conf = SparkConf().setMaster('spark://spark-master:7077').setAppName('Transformations')
sc = SparkContext(conf=conf)
```

```python
file = sc.textFile('/opt/bitnami/spark/data/u.data')
fr = sc.textFile('/opt/bitnami/spark/data/fakefriends.csv')
```

### Apply Transformations to RDD

#### map
Return a new distributed dataset formed by passing each element of the source through a function func. 

```python
ratings = file.map(lambda x: int(x.split()[2]))
ratings.collect()
```

#### filter
Return a new dataset formed by selecting those elements of the source on which func returns true. 

```python
f_r = ratings.filter(lambda r: r if r==5 else None)
f_r.collect()
```

#### FlatMap
Similar to map, but each input item can be mapped to 0 or more output items (so func should return a Seq rather than a single item). 

```python
lines = file.flatMap(lambda x: x.split('\n'))
lines.take(10)
```

#### distinct
Return a new dataset that contains the distinct elements of the source dataset.

```python
rating_sum = ratings.distinct()
rating_sum.collect()
```

### Key Value RDD operations

#### reduceByKey
When called on a dataset of (K, V) pairs, returns a dataset of (K, V) pairs where the values for each key are aggregated using the given reduce function func, which must be of type (V,V) => V. Like in groupByKey, the number of reduce tasks is configurable through an optional second argument. 

```python
ratings_a = ratings.map(lambda x: (x,1))
ratings_a.reduceByKey(lambda a,b: a+b).collect()
```

```python
ratings_a = ratings.map(lambda x: (x,1))
ratings_a.countByKey()
```

```python
ratings_a.sortByKey().collect()
```

#### mapValues(f)
Pass each value in the key-value pair RDD through a map function without changing the keys; this also retains the original RDD’s partitioning.


```python
def get_age_friends(line):
    line = line.split(',')
    age = int(line[2])
    num_friends = int(line[3])
    return (age, num_friends)

step_on = fr.map(get_age_friends)
step_tw = step_on.mapValues(lambda x: (x, 1))
total_by_age = step_tw.reduceByKey(lambda a, b: (a[0]+b[0], a[1]+b[1]))

total_by_age.collect()
```

```python
avg_age = total_by_age.mapValues(lambda x: x[0]/x[1])
avg_age.collect()
```
